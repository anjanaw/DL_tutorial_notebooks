```python
import numpy as np
import matplotlib.pyplot as plt

# Convert x values (exclusion ratio) to percentage
refer = np.array([0, 20, 40, 60])  # Now in percentage

# Convert y values (ROUGE-L scores) to percentage
avg_rougeL_by_refer = np.array([0.8451, 0.8606, 0.8697, 0.9110]) * 100  # Vector-MoLoRA
avg_rougeL_by_refer_2 = np.array([0.8313, 0.8308, 0.8593, 0.9071]) * 100  # LoRA
avg_rougeL_by_refer_3 = np.array([0.8352, 0.8439, 0.8645, 0.9157]) * 100  # MoRA

plt.figure(figsize=(8, 6))
ax = plt.gca()

# Plot the data with markers
ax.plot(refer, avg_rougeL_by_refer, marker='o', markersize=5, label='Vector-MoLoRA (Ours)')
ax.plot(refer, avg_rougeL_by_refer_2, marker='s', markersize=5, label='LoRA')
ax.plot(refer, avg_rougeL_by_refer_3, marker='^', markersize=5, label='MoRA')

# Format x-axis tick labels with "%" sign
ax.set_xticks(refer)
ax.set_xticklabels([f"{int(x)}%" for x in refer])  # Convert each tick label to percentage format

# Labels and title
ax.legend(fontsize=15)
ax.set_ylabel('Performance (Rouge-L)', fontsize=15)
ax.set_xlabel("Refer to Clinicians\n(Exclusion Ratio of Highly Uncertain Samples)", fontsize=15)
ax.set_title('Risk Coverage Curve', fontsize=16)

# Format tick labels
ax.tick_params(axis='both', labelsize=14)

# Display grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
# Save the figure
plt.savefig("33-Risk_coverage_curve.pdf", bbox_inches='tight', dpi=1500)
plt.close()
```
