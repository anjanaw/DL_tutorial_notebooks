```python
from glob import glob
import SimpleITK as sitk
import numpy as np

#endonasal_mri_batch2/mri_001/mri_001_mask.nii.gz

mask_path_all = glob ('endonasal_mri_batch2/*/*_mask.nii.gz')
dim_x = []
dim_y = []
dim_z = []
for mask_path in mask_path_all:
    # Load the NIfTI image
    mask_mutliclass = sitk.ReadImage(mask_path)
    mask_mutliclass_array = sitk.GetArrayFromImage(mask_mutliclass)
    dim_x.append(mask_mutliclass_array.shape[0])
    dim_y.append(mask_mutliclass_array.shape[1])
    dim_z.append(mask_mutliclass_array.shape[2])

print('dim_x (min, max):', np.min(dim_x), np.max(dim_x))
print('dim_y (min, max):', np.min(dim_y), np.max(dim_y))
print('dim_z (min, max):', np.min(dim_z), np.max(dim_z))
```

Center cropping: 256x256x256

```python
from glob import glob
import SimpleITK as sitk
import numpy as np

# Path to MRI segmentation masks
mask_path_all = glob('endonasal_mri_batch2/*/*_mask.nii.gz')

# Desired crop size
CROP_SIZE = (256, 256, 256)

def center_crop_3d(volume, crop_size):
    """
    Center crop a 3D volume to the specified crop size.

    Args:
        volume (np.ndarray): Input 3D volume.
        crop_size (tuple): Desired output size (x, y, z).

    Returns:
        np.ndarray: Center cropped 3D volume.
    """
    x, y, z = volume.shape
    crop_x, crop_y, crop_z = crop_size

    start_x = max(0, (x - crop_x) // 2)
    start_y = max(0, (y - crop_y) // 2)
    start_z = max(0, (z - crop_z) // 2)

    end_x = start_x + crop_x
    end_y = start_y + crop_y
    end_z = start_z + crop_z

    cropped_volume = volume[start_x:end_x, start_y:end_y, start_z:end_z]
    return cropped_volume

# Track the original and cropped dimensions
original_dims = []
cropped_dims = []

for mask_path in mask_path_all:
    # Load the NIfTI image
    mask_multiclass = sitk.ReadImage(mask_path)
    mask_multiclass_array = sitk.GetArrayFromImage(mask_multiclass)
    total_pixel_cls1 = np.sum(mask_multiclass_array==1)
    total_pixel_cls2 = np.sum(mask_multiclass_array==2)
    print('mask path:', mask_path)
    print('before cropped:', mask_multiclass_array.shape, 'classes inside:', np.unique(mask_multiclass_array),'num of pixels (cls1, cls2):', total_pixel_cls1, total_pixel_cls2)

    # Perform center cropping
    cropped_mask = center_crop_3d(mask_multiclass_array, CROP_SIZE)
    total_pixel_cls1 = np.sum(cropped_mask==1)
    total_pixel_cls2 = np.sum(cropped_mask==2)
    print('after center cropped:', cropped_mask.shape, 'classes inside:', np.unique(cropped_mask),'num of pixels (cls1, cls2):', total_pixel_cls1, total_pixel_cls2,'\n')


```
